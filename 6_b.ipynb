### Day 6 - Part B
Math homework but read the numbers vertically

```python
import copy
import math

from util.aoc_utility import *

DAY = 6
TEST_MODE = 0
SECTIONS = False

data_in = load_input(day=DAY, test_mode=TEST_MODE, sections=SECTIONS) 
```

```python
def input_to_dict(data_in):
    """Function to convert array input in rotated dictionary
    
    Args:
        data_in (arr): Array containing the original input strings
    
    Returns:
        eqs (arr): Array with elements containing "numbers" and "operation".
    """

    original = copy.deepcopy(data_in)

    #Rotate the input to read columns
    rotated = []
    first_line = True
    for line in original:
        for char_idx, char in enumerate(line):
            if first_line:
                rotated.append("")
            rotated[char_idx] += char

        first_line = False

    #Split into sections (1 per equation)
    line_width = len(rotated[0])
    blocks = []
    next_block = []
    for line in rotated:
        #Look for gap between equations
        if line == " "*line_width:
            blocks.append(next_block)
            next_block = []
        else:
            next_block.append(line)

    #Needed to add the final block
    blocks.append(next_block)

    #Convert blocks into equations
    eqs = []

    for block in blocks:
        #Set operator
        operator = block[0][-1]
        block[0] = block[0][:-1]

        #Get numbers
        numbers = [int(x.strip()) for x in block]

        eqs.append({"operator": operator, "numbers":numbers})

    return eqs
```

```python
def solve_eq(eq):
    """Solve a given equation
    
    Args:
        eq (dict): Contains an operator ("+" or "*") and numbers (array of ints)
    
    Returns:
        sol (int): Numeric solution
    """

    operator = eq["operator"]
    numbers = eq["numbers"]

    if operator == "+":
        sol = sum(numbers)
    elif operator == "*":
        sol = math.prod(numbers)

    return sol
```

```python
eqs = input_to_dict(data_in)

total_sol = sum(list(map(solve_eq, eqs)))

print(total_sol)
```
