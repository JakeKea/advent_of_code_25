### Day 7 - Part A
Laser splitting

```python
from util.aoc_utility import *

DAY = 7
TEST_MODE = 0
SECTIONS = False

data_in = load_input(day=DAY, test_mode=TEST_MODE, sections=SECTIONS)
beams = set()
beams.add(data_in[0].find("S"))
```

```python
def next_beam_layer(layer, beams):
    """Calculate the next layer of beams
    
    Args:
        layer (arr): Array for characters representing the next grid layer
        beams (set): Index (x coordinate) of beams from the previous layer

    Returns:
        next_beams (set): Index of beams for the next grid layer
        split_count (int): Number of splitter hit by beams
    """

    next_beams = set()
    split_count = 0

    for idx in beams:
        #Hits a splitter
        if layer[idx] == "^":
            next_beams.add(idx-1)
            next_beams.add(idx+1)
            split_count += 1
        #Empty space
        else:
            next_beams.add(idx)

    #Remove any beams outside the grid (not technically needed for part a)
    if -1 in next_beams:
        next_beams.remove(-1)
    if len(layer) in next_beams:
        next_beams.remove(len(layer))

    return next_beams, split_count
```

```python
next_beams = beams
split_count = 0

for line in data_in[1:]:
    next_beams, new_splitters = next_beam_layer(line, next_beams)
    split_count += new_splitters

print(split_count)
```
