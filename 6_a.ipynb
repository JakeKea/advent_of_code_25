```python
import copy
import math

from util.aoc_utility import *

DAY = 6
TEST_MODE = 0
SECTIONS = False

data_in = load_input(day=DAY, test_mode=TEST_MODE, sections=SECTIONS) 
```

```python
def input_to_dict(data_in):
    """Function to convert array input in rotated dictionary
    
    Args:
        data_in (arr): Array containing the original input strings
    
    Returns:
        eqs (arr): Array with elements containing "numbers" and "operation".
    """

    #Split original input into arrays
    original = copy.deepcopy(data_in)

    #Turn into elements
    original = [x.split(" ") for x in original]
    #Remove whitespace
    original = [list(filter(lambda y: y != "", x)) for x in original]
    #Turn elements into integers
    original = [list(map(int, x)) for x in original[:-1]] + [original[-1]]

    #Rotate list
    rotated = list(zip(*original))

    #Convert to dictionar
    eqs = []
    for line in rotated:
        eqs.append({"operator":line[-1], "numbers":list(line[:-1])})

    return eqs
```

```python
def solve_eq(eq):
    """Solve a given equation
    
    Args:
        eq (dict): Contains an operator ("+" or "*") and numbers (array of ints)
    
    Returns:
        sol (int): Numeric solution
    """

    operator = eq["operator"]
    numbers = eq["numbers"]

    if operator == "+":
        sol = sum(numbers)
    elif operator == "*":
        sol = math.prod(numbers)

    return sol
```

```python
eqs = input_to_dict(data_in)

total_sol = sum(list(map(solve_eq, eqs)))

print(total_sol)
```
