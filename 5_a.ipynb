### Day 5 - Part A
Which ids in section 2 of the input land in a range from section 1

```python
from util.aoc_utility import *

DAY = 5
TEST_MODE = 0
SECTIONS = True

data_in = load_input(day=DAY, test_mode=TEST_MODE, sections=SECTIONS)
valid_ranges = data_in[0]
ids_to_check = [int(x) for x in data_in[1]]
```

```python
def check_validility(valid_ranges, id):
    """Checks if a given id is in any of the valid ranges
    
    Args: 
        valid_ranges (arr): Array of ranges in the form of "X-Y"
        id (int): Id to check
    
    Returns:
        validility (bool): True if id exists in any valid range
    """

    for range in valid_ranges:
        limits = [int(x) for x in range.split("-")]
        if limits[0] <= id and id <= limits[1]:
            return True
        
    return False
```

```python
valid_ids = 0
for id in ids_to_check:
    if check_validility(valid_ranges, id):
        valid_ids += 1

print(valid_ids)
```
