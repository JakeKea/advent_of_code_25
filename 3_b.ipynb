### Day 3 - Part B
Maximise joltage. Joltage is now with 12 digits instead of 2

```python
from util.aoc_utility import *

DAY = 3
TEST_MODE = 0
SECTIONS = False

data_in = load_input(day=DAY, test_mode=TEST_MODE, sections=SECTIONS)
banks = [[int(z) for z in y] for y in [list(x) for x in data_in]]
```

```python
def solve_bank(bank, n=2):
    """For a given bank, find and return the maximum joltage.
    (A joltage is the biggest n digit number made using the digits in the bank without reordering the bank)

    Args:
    bank (arr): Array of single digit numbers represent the bank
    n (int): Number of digits to determine the joltage

    Returns:
    max_joltage: Biggest n digit number that can made from the bank digits without changing the order
    """

    cur_index = -1
    max_joltage = ""

    for n_i in range(1, n):
        bank_slice = bank[cur_index+1:-(n-n_i)]
        next_digit = max(bank_slice)
        cur_index += bank_slice.index(next_digit) + 1
        max_joltage += str(next_digit)

    #Handle last digit outside the for loop as bank[cur_index+1:-0] gives an empty list
    max_joltage += str(max(bank[cur_index+1:]))

    return int(max_joltage)
```

```python
total_joltage = 0
for bank in banks:
    total_joltage += solve_bank(bank, n=12)

print(total_joltage)
```
