### Day 3 - Part A
Maximise joltage

```python
from util.aoc_utility import *

DAY = 3
TEST_MODE = 0
SECTIONS = False

data_in = load_input(day=DAY, test_mode=TEST_MODE, sections=SECTIONS)
banks = [[int(z) for z in y] for y in [list(x) for x in data_in]]
```

```python
def solve_bank(bank):
    """For a given bank, find and return the maximum joltage.
    (A joltage is the biggest 2 digit number made using the digits in the bank without reordering the bank)

    Args:
    bank (arr): Array of single digit numbers represent the bank

    Returns:
    max_joltage: Biggest 2 digit number that can made from the bank digits without changing the order
    """

    #First digit can't be the last one
    first_digit = max(bank[:-1])

    max_joltage = 0
    for idx, digit in enumerate(bank[:-1]):
        if digit == first_digit:
            second_digit = max(bank[idx+1:])
            joltage = digit*10 + second_digit

            if joltage > max_joltage:
                max_joltage = joltage

    return max_joltage
```

```python
total_joltage = 0
for bank in banks:
    total_joltage += solve_bank(bank)

print(total_joltage)
```
