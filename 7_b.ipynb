Laser splitting with time nonsense

```python
from util.aoc_utility import *

DAY = 7
TEST_MODE = 0
SECTIONS = False

data_in = load_input(day=DAY, test_mode=TEST_MODE, sections=SECTIONS)
start = (data_in[0].find("S"), 0)
```

```python
def splitter_possibilities(grid, splitter_map, splitter_id):
    """Get all possibilities for a given splitter
    
    Args: 
        grid (arr): 2D Array of the input
        splitter_map (dict): Dictionary mapping each splitter to possibilities
        splitter_id (tuple):  Coordinates of the splitter
    
    Returns:
        pos (int): Number of possibilities
    """

    idx_x = splitter_id[0]
    idx_y = splitter_id[1]

    #Either hits the bottom or an already solved splitter
    for idx_dy, line in enumerate(grid[idx_y+1:]):
        if line[idx_x] != ".":
            return splitter_map[(idx_x, idx_y + idx_dy + 1)]
        
    return 1
```

```python
def solve_grid(grid):
    """From the bottom up, get the number of possibilities from each splitter
    
    Args:
        grid (arr): 2D Array of the input
            
    Returns:
        splitter_map (dict): Dictionary mapping each splitter to possibilities    
    """
    splitter_map = {}
    
    #Iterate through the grid backwards
    for idx_y in range(len(grid)-1, -1, -1):
        line = grid[idx_y]

        for idx_x, char in enumerate(line):
            if char != ".":
                pos_l = splitter_possibilities(grid, splitter_map, (idx_x-1, idx_y))
                pos_r = splitter_possibilities(grid, splitter_map, (idx_x+1, idx_y))
                splitter_map[(idx_x, idx_y)] = pos_l + pos_r

    print(splitter_map[start])
```

```python
solve_grid(data_in)
```
